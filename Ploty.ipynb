# **Atividade com Pandas**

```python
import pandas as pd
import plotly.express as px
df = pd.read_csv('https://raw.githubusercontent.com/plotly/datasets/master/gapminder2007.csv')
df = px.data.gapminder().query("continent== 'Americas' and year >= 2000 and pop > 40000000")
#df.head()
px.line(df,x='country',y='lifeExp',title='Expetativa de Vida', color="country", text='pop', hover_name='pop')
```

# **csv simples**

```python
df = pd.read_csv('https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/viagens.csv', sep=";")
#df.head()
df_meses_ano = df.groupby(['ano','mes'], sort=False)['viagens'].mean().reset_index()

fig = px.line(df_meses_ano, x='mes', y='viagens', color='ano', text='viagens', title='Viagens por mês e ano')
fig.show()
```

# **Grafico de Barras**

```python
import numpy as np

df = pd.read_csv('https://raw.githubusercontent.com/profivan-ai/cdb-Python/refs/heads/main/arquivos/compras.csv', sep=";")
df
#df_dia=df.groupby("dia",sort=False).agg({'valor': np.sum}).reset_index()
#df_dia.count()
#df_dia
#px.bar(df_dia, x='dia', y='valor', title='Gasto por dia da semana', color='dia')
```

```python
df_dia=df.groupby(["dia","sexo"],sort=False).agg({'valor': np.sum}).reset_index()
px.bar(df_dia, x='dia', y='valor', title='Gasto por Publico (M/F)', color='sexo', barmode='group', text='valor')
```

```python
df_dia=df.groupby(["dia","sexo"],sort=False).agg({'valor': np.sum}).reset_index()
px.bar(df_dia, x='valor', y='dia', title='Gasto por Publico (M/F)', color='sexo', barmode='group', text='valor', orientation='h')
```

```python
df_dia=df.groupby(["dia","sexo"],sort=False).agg({'valor': np.sum}).reset_index()
#df.head
px.bar(df_dia, x='dia', y='valor', title='Gasto por Publico (M/F)', color='sexo', barmode='group', text='valor')
```

```python
df_agrupado_dia = df.groupby("dia", sort=False).agg({'valor': 'sum'}).reset_index()
df_agrupado_dia
```

```python
px.pie(df_agrupado_dia, values="valor", names='dia', title='Gasto por Dia', color_discrete_sequence=px.colors.sequential.RdBu)
```

```python
import plotly.graph_objects as go

labels = ['Oxygem', 'Hydrogen', 'Carbon_Dioxide', 'Nitrogen']
values = [4500, 2500, 1053, 500]

#pull is given as a fraction of the pie radius
fig = go.Figure(data=[go.Pie(labels=labels, values=values, pull=[0, 0, 0, 0.2, 0])])
fig.show()
```

```python
import plotly.graph_objects as go

labels = ['Oxygem', 'Hydrogen', 'Carbon_Dioxide', 'Nitrogen']
values = [4500, 2500, 1053, 500]
colors = ['gold', 'mediumturquoise', 'darkorange', 'lightgreen']

#pull is given as a fraction of the pie radius
fig = go.Figure(
  data=[
     go.Pie(
         labels=labels,
         values=values,
         textfont_size=20,
         marker=dict(colors=colors, pattern=dict(shape=[".", "x", "+", "-"]))
     )
   ]
 )
fig.show()
```
